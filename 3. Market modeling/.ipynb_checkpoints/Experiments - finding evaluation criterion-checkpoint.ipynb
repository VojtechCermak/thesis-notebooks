# 1. Explaining volatility using tweet count


In the first experiment, we explore the prospect of tweet counts predicting volatility. From visualisation of plot it is obvious that it should be possible

## 1.1.1 Tweet count - Google

```python
np.random.seed(7)
model = LinearRegression()
iterations = 30
results_model = []
results_benchmark = []

for i in range(iterations):
    market_day = market_GOOG['Close'].at_time('16:00')
    market_day.index = market_day.index.date
    df = pd.DataFrame(tweets_GOOG_day['tweet_count']).join(market_day, how = 'inner')
    returns = df['Close'].pct_change()
    df['Volatility'] = returns*returns
    df = df.dropna().sample(frac=1)

    ### Label + features
    label = df['Volatility']
    features = pd.DataFrame(df['tweet_count'])
    
    ### MSE for model
    pred_nolag = cross_val_predict(model, features, label, cv=5, n_jobs=1, verbose=0)
    MSE_nolag = mean_squared_error(label, pred_nolag)
    results_model.append(MSE_nolag)

    ### MSE for baseline
    mean = pd.DataFrame(label.index)
    mean['mean'] = label.mean()
    MSE_benchmark = mean_squared_error(label, mean['mean'])
    results_benchmark.append(MSE_benchmark)
    
print('MSE - Tweet count without lags:  ' + str(np.mean(results_model)))
print('MSE - Mean of volatility :       ' + str(np.mean(results_benchmark)))

plot_df_GOOG = pd.DataFrame(label)
plot_df_GOOG['Prediction'] = pred_nolag
plot_df_GOOG['Mean of volatility'] = label.mean()
plot_df_GOOG = plot_df_GOOG.sort_index()
plot_df_GOOG.plot(title = 'Google - Predicting volatility using Tweet count', figsize = (12, 5), style=style1)
```

We can safely conclude, that tweet count is not very good predictor of volatility by it self. However, as can be seen from the plot, it does explain volatility spikes well. As so, it can be used for augumentation of GARCH or other volatility prediction methods

# 2. Modeling daily volatility using aggregated tweets

next step is to model volatility using vectorized tweets

## 2.1. Apple

```python
%%time
np.random.seed(7)
iterations = 30
model = LinearRegression()
results_model = []
results_benchmark = []

for i in range(iterations):
    market_day = market_AAPL['Close'].at_time('16:00')
    market_day.index = market_day.index.date

    market = market_day.shift(0)
    df = tweets_AAPL_day.join(market, how = 'inner')
    df['RET'] = df['Close'].pct_change()
    df['Volatility'] = df['RET']*df['RET']
    df = df.dropna()
    df = df.sample(frac=1)

    # Label + Features
    label = df['Volatility']
    vec = TfidfVectorizer(preprocessor=lambda x: x, tokenizer=lambda x: x, sublinear_tf = True, use_idf=True)
    features = vec.fit_transform(df['text'])

    # MSE calculation
    pred_text = cross_val_predict(model, features, label, cv=5, n_jobs=1, verbose=0)
    mse_model = mean_squared_error(label, pred_text)
    mse_mean = mean_squared_error(label, np.full(len(label), label.mean()))
    results_model.append(mse_model)
    results_benchmark.append(mse_mean)

print('MSE of model: ' + str(np.mean(results_model)))
print('Benchmark:    ' + str(np.mean(results_benchmark)))

AAPL_plot_df = pd.DataFrame(label)
AAPL_plot_df['Prediction'] = pred_text
AAPL_plot_df['Mean of volatility'] = label.mean()
AAPL_plot_df = AAPL_plot_df.sort_index()
AAPL_plot_df.plot(title = 'Apple - Predicting volatility using TF-IDF', figsize = (12, 5), style=style1)
```

# Predicting noise

Model is unable to model the spike directly. But, it learns to connect underlying twitter feed to that spike. In twitter, there are persistent underlying topics, that can last for some time, for example the ecryption dispute over FBI or Warren Buffet purchase. 

- If the spike is not in train set, model will learn nothing.
- If the spike is in the train set, model will link the underlying topic to the spike and all other observation with the underlying will have increased predicted values.


To test this, we set following hypotheses:

- Tweets are unable to out of sample predict noise without spikes
- Tweets are unable to out of sample predict noise with single spike
- Tweets are able to out of sample predict noise with cluster of spikes



## 1. Noise without spikes

## 3.3 Expanding window validation

```python
market_day = market_GOOG['Close'].at_time('16:00')
market_day.index = market_day.index.date
noise = pd.Series(np.random.normal(size = len(market_day)), index = market_day.index).rename('Normal sample')
noise[176] = 15
noise[177] = 18
noise[178] = 20
noise[179] = 18
noise[180] = 15
features, label = tweets_GOOG_day.align(noise, axis=0, join = 'inner')
label = label*label
vec = TfidfVectorizer(preprocessor=lambda x: x, tokenizer=lambda x: x, sublinear_tf = True, use_idf=True)
features = vec.fit_transform(features['text'])


model = LinearRegression()

window = 1
rolling_results_1w = np.ones(0)
rolling_label_1w = np.ones(0)
for i in range(5, 100, window):
    model.fit(features[:i], label[:i])
    pred = model.predict(features[i:i+window])
    rolling_results_1w = np.append(rolling_results_1w, pred)
    rolling_label_1w = np.append(rolling_label_1w, label[i:i+window])
    
window = 10    
rolling_results_10w = np.ones(0)
rolling_label_10w = np.ones(0)
for i in range(5, 100, window):
    model.fit(features[:i], label[:i])
    pred = model.predict(features[i:i+window])
    rolling_results_10w = np.append(rolling_results_10w, pred)
    rolling_label_10w = np.append(rolling_label_10w, label[i:i+window])
    
fig, ax = plt.subplots(1,2, figsize = (14, 5))
window_1 = pd.DataFrame({'Prediction' : rolling_results_1w, 'Volatility' : rolling_label_1w})
window_10 = pd.DataFrame({'Prediction' : rolling_results_10w, 'Volatility' : rolling_label_10w})
window_1.plot(ax = ax[0], style = style1, title= 'Expanding window validation with window size 1')
window_10.plot(ax = ax[1], style = style1, legend = None, title= 'Expanding window validation with window size 10')
ax[0].legend(loc='upper center', bbox_to_anchor=(1.1, -0.1), fancybox=True, shadow=True,  ncol=5)
ax[0].set_ylabel("Volatility")
ax[1].set_ylabel("Volatility")
fig.savefig('figures\\experiment_rolling_failure.png')
```

## Offsets

```python
%%time
AAPL_price = market_AAPL['Close'].between_time('16:00', '16:00').rename('AAPL_market')
AAPL_price.index = AAPL_price.index.date
GOOG_price = market_GOOG['Close'].between_time('16:00', '16:00').rename('GOOG_market')
GOOG_price.index = GOOG_price.index.date
prices = pd.concat([GOOG_price, AAPL_price], axis = 1)

AAPL_cross_model= []
AAPL_cross_benchmark = []
GOOG_cross_model= []
GOOG_cross_benchmark = []

for i in list(range(-120, 120, 5)):
    #print(i)
    prices = pd.concat([GOOG_price, AAPL_price], axis = 1)
    prices.index = prices.index + pd.DateOffset(days = i)
    df = pd.concat([tweets_AAPL_day['text'].rename('AAPL'), tweets_GOOG_day['text'].rename('GOOG')], axis = 1)
    df = df.join(prices, how = 'inner')

    AAPL_ret = df['AAPL_market'].pct_change()
    GOOG_ret = df['GOOG_market'].pct_change()
    df['AAPL_vol'] = AAPL_ret*AAPL_ret
    df['GOOG_vol'] = GOOG_ret*GOOG_ret

    df = df.dropna()
    df = df.sample(frac=1)

    label_GOOG = df['GOOG_vol']
    label_AAPL = df['AAPL_vol']
    
    # Features
    text_combined = pd.concat([df['AAPL'], df['GOOG']])
    vec = TfidfVectorizer(preprocessor=lambda x: x, tokenizer=lambda x: x, sublinear_tf = True, use_idf=True)
    vec_combined = vec.fit_transform(text_combined)
    vec_AAPL = vec_combined[:len(df)]
    vec_GOOG = vec_combined[len(df):]

    # Model
    model = LinearRegression()
    
    
    ## MODELING - Train on apple, predict Google
    model.fit(vec_AAPL, label_AAPL)
    pred_text = model.predict(vec_GOOG)
    GOOG_mse_model = mean_squared_error(label_GOOG, pred_text)
    GOOG_mse_mean = mean_squared_error(label_GOOG, np.full(len(label_GOOG), label_GOOG.mean()))
    GOOG_cross_model.append(GOOG_mse_model)
    GOOG_cross_benchmark.append(GOOG_mse_mean)
    
    ## MODELING - Train on Google, predict Apple
    model.fit(vec_GOOG, label_GOOG)
    pred_text = model.predict(vec_AAPL)
    AAPL_mse_model = mean_squared_error(label_AAPL, pred_text)
    AAPL_mse_mean = mean_squared_error(label_AAPL, np.full(len(label_AAPL), label_AAPL.mean()))
    AAPL_cross_model.append(AAPL_mse_model)
    AAPL_cross_benchmark.append(AAPL_mse_mean)
```

```python
AAPL_cross_results = pd.DataFrame({'Baseline MSE' : AAPL_cross_benchmark, 'Cross prediction MSE' : AAPL_cross_model})
AAPL_cross_results.index = (AAPL_cross_results.index - 24)*5
fig, ax = plt.subplots(1,2, figsize = (14, 5))
AAPL_cross_results.plot(title = 'Apple - MSE of cross predicitonon shifted label', ax = ax[0], style=style1)
ax[0].set_xlabel('Days shifted')
ax[0].set_ylabel('MSE')
ax[0].plot([0, 0], [0.00000004, 0.0000007], 'k:', lw=1)

AAPL_share = (AAPL_cross_results['Cross prediction MSE']) / (AAPL_cross_results['Baseline MSE'])
AAPL_share.plot(title = 'MSE model / MSE Baseline', ax = ax[1], legend = None, style=style1)
ax[1].set_xlabel('Days shifted')
ax[1].set_ylabel('% MSE of Baseline')
ax[1].plot([-120, 120], [1, 1], 'k:', lw=1)
fig.savefig('figures\\experimetns_cross_offsets_apple.png')
```
